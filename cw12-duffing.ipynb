#### Problem 1
$F = 0.18$, plot $x(t)$ for $t\in[0,2\pi\, 50]$, with $x(0) = -0.9$ and $y(0) = 0$

```python
t,x,y = sm.rk4(0, 50, -0.9, 0, 0.25, 0.18)
```

```python
sm.plot(t,x, "x(t)")
sm.parametric(x,y, "Parametric curve")
sm.scatter(x,y,50, "Poincare")
```

The x(t) graph has some irregularity, but quickly establishes a regular oscillation. The parametric curve is shaped like an egg. The poincare section shows what looks like the beginning of a curve.

#### Problem 3

$F = 0.25$, plot $x(t)$ for $t\in[0,2\pi\, 50]$ with $x(0) = 0.2$, and $y(0) = 0.1$

```python
t,x,y = sm.rk4(0,50,.2,.1,.25,.25)
```

```python
sm.plot(t,x, "x(t)")
sm.parametric(x,y, "Parametric curve")
sm.scatter(x,y,50, "Poincare")
```

The x(t) curve oscillates before jumping down to oscillate before jumping back up and oscillating regularly. The parametric curve reminds me of cell division. The Poincare section looks relatively linear.

##### Changing $x_0$ from 0.2 to 0.201:

```python
t,x,y = sm.rk4(0,50,.201,.1,.25,.25)
```

```python
sm.plot(t,x, "x(t)")
sm.parametric(x,y, "Parametric curve")
sm.scatter(x,y,50, "Poincare")
```

When the initial value of x is changed by only .001, the x(t) graph's period of regular oscillation is around $\dot{x} = -1$ where before it oscillated around $\dot{x} = 1$. The Poincare section looks similar in shape.
